```python
import pandas as pd
from bs4 import BeautifulSoup

data_file = 'cities.csv'

df = pd.read_csv(data_file)

data_clean = df[['City_ID', 'City', 'Country', 'Max Temp', 'Humidity', 'Cloudiness', 'Wind Speed', 'Lat', 'Lng', 'Date']]

cities = data_clean.set_index('City_ID')

cities.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import date

#export cities csv file and display it as a dataset
cities_data= pd.DataFrame(cities)

lat = cities_data["Lat"]
maxtemps = cities_data["Max Temp"]
humidity = cities_data["Humidity"]
cloudiness = cities_data["Cloudiness"]
windspeed = cities_data["Wind Speed"]
date = cities_data["Date"]

city_data = df.set_index('City_ID')

cities.head()
 
```

```python
#Begin creating scatter plot using maps to get lat & lon
#date = date.today().strftime("%m/%d/%y")
date = '01/05/17'

plt.scatter(cities_data["Lat"],cities_data["Max Temp"], facecolor = "steelblue",edgecolor = "black")
 
plt.title(f"City Latitude Vs Max Temperature ({date})")
plt.xlabel("Laitude")
plt.ylabel("Max Temperature")
#plot the gridline
plt.grid(linestyle='-',linewidth=1, alpha = 0.18)
plt.savefig("../Resources/City Latitude_vs_Max Temperature.png")
```
